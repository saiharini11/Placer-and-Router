```python
class TreeNode:
    #bend=0
    def __init__(self,data,cost):
        self.data=data
        self.cost=cost
        self.children=[]
        self.parent=None
        
    def add_child(self,child):
        child.parent=self
        self.children.append(child)
        #child.cost=get_cost(child)
        
    def get_cost(self,child,bend):  
        child.cost=child.cost+self.cost   
        # Check if there is any turn
        #print("self.parent:",self.parent.data)
        if(self.parent!=None):
            
            if(not(child.data[0]==self.data[0]==self.parent.data[0] or child.data[1]==self.data[1]==self.parent.data[1])):
                print("class ### self.parent:",self.parent.data)
                child.cost=child.cost+bend
        return child.cost
    def get_level(self):
        p=self
        level=0
        while(p):
            level=level+1
            p=p.parent
        return level
```

```python
def get_adjacentCell(graph,x,y,temp,r1):
    temp1=[]
    #temp1.append(y)
    temp1.append(x)
    temp1.append(y-r1)
    temp1.append(graph[y][x])
    temp.append(temp1)
    return temp
```

```python
def get_adjacencyList(r1,r2,cols,graph):
    adjacent_list=[]
    for x in range(cols):
    #for y in range(r1,r2):
        adjacent=[]
        for y in range(r1,r2):
        #for x in range(cols):
            temp=[]
            if(graph[y][x]!=-1):
                # Check adjacent cells in right side
                if(x+1<cols and graph[y][x+1]!=-1):
                    temp=get_adjacentCell(graph,x+1,y,temp,r1)
                # Check adjacent cells in upper side
                if(y-1>=r1 and graph[y-1][x]!=-1):
                    temp=get_adjacentCell(graph,x,y-1,temp,r1)
                # Check adjacent cells in left side
                if(x-1>=0 and graph[y][x-1]!=-1):
                    temp=get_adjacentCell(graph,x-1,y,temp,r1)
                # Check adjacent cells in lower side
                if(y+1<r2 and graph[y+1][x]!=-1):
                    temp=get_adjacentCell(graph,x,y+1,temp,r1)  
                if(temp==[]):
                    print("Not possible to execute") 
            adjacent.append(temp)    
        adjacent_list.append(adjacent) 
    return adjacent_list
```

```python
def get_minCost(pre_cost):
    min_cost=pre_cost[0]
    for i in range(len(pre_cost)):
        if(pre_cost[i]<min_cost):
            min_cost=pre_cost[i]
    return min_cost
```
